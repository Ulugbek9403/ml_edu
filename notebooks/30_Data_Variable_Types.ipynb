- –∑–∞–ø–æ–ª–Ω–µ–Ω–∏–µ —Å—Ä–µ–¥–Ω–∏–º;
- –∑–∞–ø–æ–ª–Ω–µ–Ω–∏–µ –∫–æ–Ω—Å—Ç–∞–Ω—Ç–æ–π (–æ–±—ã—á–Ω–æ, —Å–∏–ª—å–Ω–æ –ø—Ä–∏–≤—è–∑–∞–Ω–æ –∫ –±–∏–∑–Ω–µ—Å-–∑–∞–¥–∞—á–µ);
- –∑–∞–ø–æ–ª–Ω–µ–Ω–∏–µ –º–µ–¥–∏–∞–Ω–Ω—ã–º –∑–Ω–∞—á–µ–Ω–∏–µ–º.

sklearn —Ä–µ–∞–ª–∏–∑—É–µ—Ç –ø—Ä–æ—Å—Ç–æ–π –∫–ª–∞—Å—Å –¥–ª—è –∑–∞–ø–æ–ª–Ω–µ–Ω–∏—è —Å–æ –≤—Å–µ–º–∏ –ø–µ—Ä–µ—á–∏—Å–ª–µ–Ω–Ω—ã–º–∏ –≤–∞—Ä–∏–∞–Ω—Ç–∞–º–∏ [SimpleImputer](https://scikit-learn.org/stable/modules/generated/sklearn.impute.SimpleImputer.html):

```python
# –î–æ–±–∞–≤–∏–º –ø—Ä–æ–ø—É—Å–∫–æ–≤ –≤ –¥–∞–Ω–Ω—ã–µ
df_data_nans = df_data.copy()

df_data_nans.iloc[[2, 8, 15, 20, 60, 100]] = np.nan

df_data_nans.isna().sum()
```

```python
df_data_nans.head(10)
```

```python
sns.displot(df_data_nans, height=7, aspect=1.5)
plt.show()
```

```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy="mean")

imputer.fit(df_data_nans)
imputer.statistics_
```

–ö–∞–∫ –≤–∏–¥–Ω–æ, –Ω–∞ –æ—Å–Ω–æ–≤–µ –≤—ã–±—Ä–∞–Ω–Ω–æ–π —Å—Ç—Ä–∞—Ç–µ–≥–∏–∏ —Ä–∞—Å—Å—á–∏—Ç—ã–≤–∞–µ—Ç—Å—è –ø–æ–∫–∞–∑–∞—Ç–µ–ª—å, –∞ –∑–∞—Ç–µ–º –æ–Ω –∑–∞–ø–æ–ª–Ω—è–µ—Ç—Å—è –≤ –¥–∞–Ω–Ω—ã—Ö:

```python
data_filled_na = df_data_nans.copy()

data_filled_na["num_var_filled"] = imputer.transform(df_data_nans[["num_var"]])

data_filled_na.head(10)
```

–ó–Ω–∞—á–µ–Ω–∏–µ –ø–æ–¥—Å—Ç–∞–≤–∏–ª–æ—Å—å –ø—Ä—è–º–∏–∫–æ–º –≤ –º–µ—Å—Ç–∞, –≥–¥–µ —Ä–∞–Ω–µ–µ –±—ã–ª–∏ NaN. –°—Ä–∞–±–æ—Ç–∞–ª–æ!

–î–∞–≤–∞–π—Ç–µ –≥–ª—è–Ω–µ–º –Ω–∞ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ:

```python
sns.displot(data_filled_na, height=7, aspect=1.5)
plt.show()
```

## –ö–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã–µ –ø–µ—Ä–µ–º–µ–Ω–Ω—ã–µ

–í–æ—Ç —Ç—É—Ç –Ω–∞—á–∏–Ω–∞–µ—Ç—Å—è —Å–∞–º–æ–µ –∏–Ω—Ç–µ—Ä–µ—Å–Ω–æ–µ!

–í –Ω–µ–∫–æ—Ç–æ—Ä—ã—Ö —Å–ª—É—á–∞—è—Ö –≤–∞–º –ø–æ–≤–µ–∑—ë—Ç, –∏ —É –≤–∞—Å —É–∂–µ –±—É–¥—É—Ç –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã–µ –ø–µ—Ä–µ–º–µ–Ω–Ω—ã–µ –≤ –≤–∏–¥–µ —á–∏—Å–µ–ª (—Å–∫–æ—Ä–µ–µ –≤—Å–µ–≥–æ —Ü–µ–ª–æ—á–∏—Å–ª–µ–Ω–Ω—ã—Ö) –∏ –±–µ–∑ –ø—Ä–æ–ø—É—Å–∫–æ–≤!

–ù–æ –Ω–µ —Å–ø–µ—à–∏—Ç–µ –ø—Ä–æ–ø—É—Å–∫–∞—Ç—å —ç—Ç—É –≥–ª–∞–≤—É. –í —Ö—É–¥—à–µ–º —Å–ª—É—á–∞–µ –∫–∞—Ç–µ–≥–æ—Ä–∏–∏ –±—É–¥—É—Ç –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω—ã —Å—Ç—Ä–æ–∫–∞–º–∏ –∏ –∏–º–µ—Ç—å –ø—Ä–æ–ø—É—Å–∫–∏. –í–æ—Ç —Ç–æ–≥–¥–∞ –≤—ã –∏ –≤—Å–ø–æ–º–Ω–∏—Ç–µ, —á–µ–º—É –º—ã —Ç—É—Ç —Å–µ–π—á–∞—Å –Ω–∞—É—á–∏–º—Å—è!

–í —ç—Ç–æ–º –Ω–∞–º –ø–æ–º–æ–∂–µ—Ç mapping –∏–∑ —Å—Ç—Ä–æ–∫ –≤ —á–∏—Å–ª–∞. –ù–∞–ø—Ä–∏–º–µ—Ä, –º—ã –∏–º–µ–µ–º —Ç–∞–∫–∏–µ –¥–∞–Ω–Ω—ã–µ:

```python
categorical_variable = np.random.RandomState(RANDOM_SEED).choice(["Red", "Blue", "Black", "White"], size=(100,), replace=True)
df_data = pd.DataFrame({"cat_var": categorical_variable})

df_data.head(10)
```

```python
df_data.info()
```

–ü–æ–ø—Ä–æ–±—É–µ–º —Å–¥–µ–ª–∞—Ç—å –º–∞—Ç—Ä–∏—Ü—É —Å —á–µ—Ç—ã—Ä–µ–º—è –∫–æ–ª–æ–Ω–∫–∞–º–∏, –∫–∞–∂–¥–∞—è –∏–∑ –∫–æ—Ç–æ—Ä—ã—Ö –±—É–¥–µ—Ç –∏–Ω–¥–∏–∫–∞—Ç–æ—Ä–æ–º –∫–∞–∂–¥–æ–≥–æ —Ü–≤–µ—Ç–∞. –ü–µ—Ä–≤–∞—è –∫–æ–ª–æ–Ω–∫–∞ - "—è–≤–ª—è–µ—Ç—Å—è –ª–∏ –ø—Ä–∏–º–µ—Ä —Ü–≤–µ—Ç–æ–º Black?". –í—Ç–æ—Ä–∞—è - –¥–ª—è Blue –∏ —Ç.–¥.

–î–ª—è —ç—Ç–æ–≥–æ –Ω–∞–º –ø–æ–º–æ–∂–µ—Ç [OneHotEncoder](https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.OneHotEncoder.html):

```python
from sklearn.preprocessing import OneHotEncoder

nom_enc = OneHotEncoder()

categorical_columns = ["cat_var"]
encoded_data = nom_enc.fit_transform(df_data[categorical_columns]).toarray()
encoded_df = pd.DataFrame(encoded_data.astype(int), columns=nom_enc.get_feature_names_out(categorical_columns))
encoded_df["cat_var"] = df_data["cat_var"]
encoded_df.head()
```

> üî• –û–±—Ä–∞—Ç–∏—Ç–µ –≤–Ω–∏–º–∞–Ω–∏–µ, `categorical_columns` - —ç—Ç–æ –ø–µ—Ä–µ–º–µ–Ω–Ω–∞—è, –∫–æ—Ç–æ—Ä–∞—è —Ö—Ä–∞–Ω–∏—Ç —Å–ø–∏—Å–æ–∫ –∏–º—ë–Ω –ø—Ä–∏–∑–Ω–∞–∫–æ–≤, –∫–æ—Ç–æ—Ä—ã–µ —Å–æ–¥–µ—Ä–∂–∞—Ç –Ω–æ–º–∏–Ω–∞–ª—å–Ω—ã–µ –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã–µ –ø–µ—Ä–µ–º–µ–Ω–Ω—ã–µ. –≠—Ç–æ –Ω—É–∂–Ω–æ, —á—Ç–æ–±—ã –≤ —Ä–µ–∞–ª—å–Ω–æ–º –¥–∞—Ç–∞—Å–µ—Ç–µ –Ω–µ –ø–µ—Ä–µ–¥–∞—Ç—å –Ω–∞ –∫–æ–¥–∏—Ä–æ–≤–∞–Ω–∏–µ –≤—Å–µ –∫–æ–ª–æ–Ω–∫–∏, –≤ —Ç–æ–º —á–∏—Å–ª–µ –∏ —á–∏—Å–ª–æ–≤—ã–µ, –∏ —Å –¥–∞—Ç–∞–º–∏.

–í–æ—Ç —Ç–∞–∫ –Ω–µ—Å–ª–æ–∂–Ω–æ –º—ã —Å–º–æ–≥–ª–∏ –ø—Ä–µ–≤—Ä–∞—Ç–∏—Ç—å —Å—Ç—Ä–æ–∫–æ–≤—ã–µ –∫–∞—Ç–µ–≥–æ—Ä–∏–∏ –≤ —á–∏—Å–ª–∞, —á—Ç–æ —É–∂–µ –ø–æ–∑–≤–æ–ª—è–µ—Ç –æ–±—É—á–∞—Ç—å –º–æ–¥–µ–ª—å!



### –ö–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã–µ –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω—ã–µ –ø–µ—Ä–µ–º–µ–Ω–Ω—ã–µ

–ù—É, –∞ —Å –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω—ã–º–∏-—Ç–æ –º—ã –º–æ–∂–µ–º –∫–æ–¥–∏—Ä–æ–≤–∞—Ç—å —á–∏—Å–ª–∞–º–∏??

–ö–æ–Ω–µ—á–Ω–æ! –ó–¥–µ—Å—å –Ω–µ—Ç –æ–≥—Ä–∞–Ω–∏—á–µ–Ω–∏—è –Ω–∞ –Ω–µ–∑–∞–≤–∏—Å–∏–º–æ—Å—Ç—å, –Ω–æ –µ—Å—Ç—å –≤–∞–∂–Ω–æ—Å—Ç—å –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç–∏, –ø–æ—ç—Ç–æ–º—É –≤ [OrdinalEncoder](https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.OrdinalEncoder.html) –≤–∞–∂–Ω–æ –ø–µ—Ä–µ–¥–∞—Ç—å —Å–ø–∏—Å–æ–∫ –∫–∞—Ç–µ–≥–æ—Ä–∏–π –≤ —Ç–æ–º –ø–æ—Ä—è–¥–∫–µ, –≤ –∫–æ—Ç–æ—Ä–æ–º –æ–Ω–∏ –¥–æ–ª–∂–Ω—ã —Å—Ç–æ—è—Ç—å!

```python
categorical_variable = np.random.RandomState(RANDOM_SEED).choice(["Bad", "Good", "Nice", "Very well"], size=(100,), replace=True)
df_data = pd.DataFrame({"cat_var": categorical_variable})
df_data["cat_var"] = df_data["cat_var"].astype("category")

df_data.head(10)
```

```python
from sklearn.preprocessing import OrdinalEncoder

seq_enc = OrdinalEncoder(categories=[["Bad", "Good", "Nice", "Very well"]])

categorical_columns = ["cat_var"]

encoded_data = seq_enc.fit_transform(df_data[categorical_columns])

encoded_df = pd.DataFrame(encoded_data.astype(int), columns=["cat_var_enc"])
encoded_df["cat_var"] = df_data["cat_var"]
encoded_df.head(10)
```

–ü—Ä–µ–∫—Ä–∞—Å–Ω–æ! –í–æ—Ç —Ç–∞–∫ –ª–µ–≥–∫–æ –∏ –∏–∑—è—â–Ω–æ –º—ã –Ω–∞—É—á–∏–ª–∏—Å—å –∫–æ–¥–∏—Ä–æ–≤–∞—Ç—å –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã–µ –ø–µ—Ä–µ–º–µ–Ω–Ω—ã–µ –ª—é–±–æ–≥–æ —Ç–∏–ø–∞!

–í–∞–∂–Ω–æ –Ω–µ –∑–∞–±—ã–≤–∞—Ç—å —Ä–∞–¥–∏ —á–µ–≥–æ –≤—Å—ë —ç—Ç–æ –¥–µ–ª–∞–µ—Ç—Å—è, –∏ —Ç–æ–≥–¥–∞ —Å—Ç–∞–Ω–æ–≤–∏—Ç—Å—è –ø–æ–Ω—è—Ç–Ω–æ, –∫–∞–∫–∏–µ –ø–æ–¥—Ö–æ–¥—ã –ª—É—á—à–µ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –≤ —Ç–æ–π –∏–ª–∏ –∏–Ω–æ–π —Å–∏—Ç—É–∞—Ü–∏–∏! –ú—ã –∂–µ —Å—Ç–∞—Ä–∞–µ–º—Å—è –ø–æ–¥–≥–æ—Ç–æ–≤–∏—Ç—å –¥–∞–Ω–Ω—ã–µ –¥–ª—è –º–æ–¥–µ–ª–∏, —á—Ç–æ–±—ã –æ–Ω–∞ –ª—É—á—à–µ –ø–æ–Ω—è–ª–∞ –¥–∞–Ω–Ω—ã–µ, –∞ –Ω–µ –∑–∞–ø—É—Ç–∞—Ç—å –µ—ë!

–ü–æ–ø–æ–ª–Ω–∏–≤ –∞—Ä—Å–µ–Ω–∞–ª, –≤–∞–∂–Ω–æ –Ω–µ –∑–∞–±—ã–≤–∞—Ç—å, —á—Ç–æ –µ—Å—Ç—å –µ—â—ë –æ–¥–Ω–∞ –±–µ–¥–∞ - –ø—Ä–æ–ø—É—Å–∫–∏ –≤ –∫–∞—Ç–µ–≥–æ—Ä–∏—è—Ö. –ò –ø—Ä–∏ —ç—Ç–æ–º —É –Ω–∞—Å –Ω–µ—Ç –≤ –∫–∞—Ç–µ–≥–æ—Ä–∏—è—Ö –Ω–∏ –º–µ–¥–∏–∞–Ω, –Ω–∏ —Å—Ä–µ–¥–Ω–∏—Ö!

> –ï—Å–ª–∏ –ø–æ —Å–µ–∫—Ä–µ—Ç—É, —Ç–æ OneHotEncoder –Ω–µ —Ç–µ—Ä–ø–∏—Ç –ø—Ä–æ–ø—É—Å–∫–æ–≤, –ø–æ—ç—Ç–æ–º—É –≤–∞–∂–Ω–æ —Å–Ω–∞—á–∞–ª–∞ –æ–±—Ä–∞–±–æ—Ç–∞—Ç—å –ø—Ä–æ–ø—É—Å–∫–∏, –∞ –∑–∞—Ç–µ–º —Ç–æ–ª—å–∫–æ –∫–æ–¥–∏—Ä–æ–≤–∞—Ç—å.

–ß—Ç–æ –∂–µ –Ω–∞–º –¥–µ–ª–∞—Ç—å —Å –∫–∞—Ç–µ–≥–æ—Ä–∏—è–º–∏, –µ—Å–ª–∏ –Ω–∞–º –Ω—É–∂–Ω–æ –∑–∞–ø–æ–ª–Ω–∏—Ç—å –ø—Ä–æ–ø—É—Å–∫–∏? –í—Å—ë –æ—á–µ–Ω—å –ø—Ä–æ—Å—Ç–æ, top-value –ø–æ–¥—Ö–æ–¥!

–ó–∞–∫–ª—é—á–∞–µ—Ç—Å—è –æ–Ω –≤ —Ç–æ–º, —á—Ç–æ –º—ã –±–µ—Ä—ë–º —Å–∞–º—É—é —á–∞—Å—Ç—É—é –∫–∞—Ç–µ–≥–æ—Ä–∏—é –∏ –µ—ë –ø–æ–¥—Å—Ç–∞–≤–ª—è–µ–º –Ω–∞ –º–µ—Å—Ç–æ –ø—Ä–æ–ø—É—Å–∫–æ–≤.

```python
df_data_nans = df_data.copy()

na_idxs = [2, 6, 65, 12, 55, 33, 98, 16]

df_data_nans.iloc[na_idxs] = np.nan
df_data_nans.isna().sum()
```

```python
fig = sns.countplot(data=df_data_nans, x="cat_var")

# –î–æ–±–∞–≤–∏–º –ø–æ–¥–ø–∏—Å–∏ –¥–ª—è –∫–∞–∂–¥–æ–≥–æ –±–∞—Ä–∞
for c in fig.containers:
    plt.bar_label(c)

plt.show()
```

–í–æ—Ç –º—ã –≤–∏–¥–∏–º —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ, –∞ —Ç–µ–ø–µ—Ä—å –æ—Å—Ç–∞—ë—Ç—Å—è –≤—ã–±—Ä–∞—Ç—å –ø—Ä–∞–≤–∏–ª—å–Ω—É—é —Å—Ç—Ä–∞—Ç–µ–≥–∏—é —É [SimpleImputer](https://scikit-learn.org/stable/modules/generated/sklearn.impute.SimpleImputer.html):

```python
imputer = SimpleImputer(strategy="most_frequent")

imputer.fit(df_data_nans)
imputer.statistics_
```

–í–∏–¥–∏—Ç–µ, Imputer –≤—ã—á–∏—Å–ª–∏–ª –Ω–∞–∏–±–æ–ª–µ–µ —á–∞—Å—Ç—É—é –∫–∞—Ç–µ–≥–æ—Ä–∏—é, –∞ —Ç–µ–ø–µ—Ä—å –ø—Ä–æ—Å—Ç–æ –ø–æ–¥—Å—Ç–∞–≤–∏—Ç –µ—ë –≤–º–µ—Å—Ç–æ NaN:

```python
df_data_filled = df_data_nans.copy()
df_data_filled["cat_var_filled"] = imputer.transform(df_data_filled[["cat_var"]]).reshape(-1)
df_data_filled.head(10)
```
